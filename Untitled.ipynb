```python
from scipy.io import wavfile  #scipy :  과학 컴퓨팅과 기술 컴퓨팅에 사용되는 자유-오픈 소스 파이썬 라이브러리이다.(위키백과)
from scipy import signal
import numpy as np     #numpy :  행렬이나 일반적으로 대규모 다차원 배열을 쉽게 처리 할 수 있도록 지원하는 파이썬의 라이브러리이다. 
                        #         numPy는 데이터 구조 외에도 수치 계산을 위해 효율적으로 구현된 기능을 제공한다.
from matplotlib import pyplot as plt 
# matplot : Matplotlib은 파이썬에서 매트랩과 유사한 그래프 표시를 가능케 하는 라이브러리다. (위키백과)
import os # os :  OS별 작업의 노출을 허용하는 핵심 Python 모듈이다. (파이썬 위키)
```

```python
wav = r'.\CantinaBand60.wav' # Original file
(file_dir, file_id) = os.path.split(wav)

print("Path : ", file_dir) # File 경로
print("Name : ", file_id)  # File 이름
```

```python
sample_rate, data = wavfile.read(wav) # sr : sampling rate, x : wave data array

print("Sample rate:{0}, data size:{1}, duration:{2} seconds".format(sample_rate,data.shape,len(data)/sample_rate))
```

```python
time = np.linspace(0, len(data)/sample_rate, len(data))
plt.figure(figsize=(20,10))
plt.plot(time, data)
plt.ylabel("Amplitude")   # y축 : 진폭
plt.xlabel("Time[s]")     # x축 : 시간
plt.title("Amplitude - Time[s]")
plt.show()
```
